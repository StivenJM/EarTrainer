```python
import pandas as pd
import os
from IPython.display import display, HTML

nombre_archivo_csv = 'musical_skills_smote1.csv'

try:
    # 2. Cargar el archivo CSV en un DataFrame de pandas
    df = pd.read_csv(nombre_archivo_csv)
    print(f"Archivo '{nombre_archivo_csv}' cargado exitosamente.")
    print("\nPrimeras 5 filas del DataFrame original:")
    display(df.head())

    # 3. Desordenar las filas del DataFrame
    df_desordenado = df.sample(frac=1).reset_index(drop=True)
    print("\nDataFrame con filas desordenadas (primeras 5 filas):")
    display(df_desordenado.head())

    # 4. Guardar el DataFrame desordenado en un nuevo archivo CSV
    nombre_archivo_desordenado = 'musical_skills_smote_final.csv'
    df_desordenado.to_csv(nombre_archivo_desordenado, index=False)
    print(f"\nDataFrame desordenado guardado en '{nombre_archivo_desordenado}'.")

except FileNotFoundError:
    print(f"Error: El archivo '{nombre_archivo_csv}' no se encontró. Por favor, asegúrate de que el nombre y la ruta son correctos.")
except Exception as e:
    print(f"Ocurrió un error inesperado: {e}")
```
